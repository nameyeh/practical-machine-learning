```python
train = pd.read_csv('.//digit-recognizer//train.csv') #read in train data
test = pd.read_csv('.//digit-recognizer//test.csv') #read in test data
train.shape
test.shape
train.head()
test.head()
```

```python
X_train = train.drop('label', 1)  
y_train = train["label"]

# Check the shape of the trainig data set array
print('Shape of X_train_data:', X_train.shape)
print('Shape of y_train_data:', y_train.shape)
#####################################################
#RANDOM FOREST with MNIST data set (no PCA)
import time
start = time.perf_counter()

RF_clf = RandomForestClassifier(
  bootstrap = True,
  n_estimators=10,
  max_features='sqrt', 
  random_state=RANDOM_SEED
)

RF_clf.fit(X_train, y_train)

#Elapsed Time to perform Random Forest
end = time.perf_counter()
time = (end - start)
timeRF1_42000 = time 

print('Runtime for Random Forest:', timeRF1_42000)
#####################################################
prediction = RF_clf.predict(test)
print(prediction[0:10])
df=pd.read_csv('.//digit-recognizer//sample_submission.csv') 
df.head(3)
df=df.drop(columns='Label')
df.head(3)
df['Label'] = prediction
df.head(3)
df.to_csv("step1_RF_submission1_split42_28_kaggle.csv", index=False)
df.tail()
```

##### 1b) REDO with 60,000/10,000 split and eval metrics

```python
%%time 
from sklearn.datasets import fetch_openml
mnist = fetch_openml('mnist_784', version=1, cache=True)
# mnist
```

```python
#Split training data (60,000) and test data (10,000)
split = 60000
X_train, X_test = mnist["data"][:split], mnist["data"][split:]
y_train, y_test = mnist["target"][:split], mnist["target"][split:]
# Check the shape of the trainig data set array
print('Shape of X_train_data:', X_train.shape, X_test.shape)
print('Shape of y_train_data:', y_train.shape, y_test.shape)
#####################################################
# Shuffle the training set so that all cross-validation folds will be similar.
shuffle = np.random.permutation(split)
X_train, y_train = X_train[shuffle], y_train[shuffle]
#####################################################
#RANDOM FOREST with MNIST data set (no PCA)
import time
start = time.perf_counter()

RF_clf = RandomForestClassifier(
  bootstrap = True,
  n_estimators=10,
  max_features='sqrt', 
  random_state=RANDOM_SEED
)

RF_clf.fit(X_train, y_train)

#####################################################
# Evaluate a score of RANDOM FOREST by cross-validation
print(cross_val_score(
  RF_clf, 
  X_test, y_test, 
  cv=10, 
  scoring='f1_macro'
))
#####################################################
#Generate cross-validated estimates for each input data point
y_pred = cross_val_predict(
  RF_clf,
  X_test, y_test,
  cv=10
)
#####################################################
# Show Precision, Recall and F1-Score
print(classification_report(y_test, y_pred))
#####################################################
# Calcurate Average F1 Score
# http://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html#sklearn.metrics.f1_score
f1score_RF_clf = f1_score(y_test, y_pred, average='macro')
f1score_RF1_clf_60000=f1score_RF_clf
f1score_RF1_clf_60000
#####################################################
#Elapsed Time to perform Random Forest
end = time.perf_counter()
time = (end - start)
timeRF1_60000 = time 

print('Runtime for Random Forest:', timeRF1_60000)
#####################################################
#Confusion Matrix for RANDOM FOREST with MNIST data set (no PCA)
cm_RF_clf = confusion_matrix(y_test, y_pred)
cm_RF_clf
#####################################################
# Show Confusion Matrix Plot for RANDOM FOREST with MNIST data set (no PCA)
plt.matshow(cm_RF_clf, cmap=plt.cm.gray)
plt.title('ConfusionMatrix_RF_Clf')
plt.show()
#####################################################
# Show Error Plot for RANDOM FOREST with MNIST data set (no PCA) 
rws = cm_RF_clf.sum(axis=1, keepdims=True)
norm = cm_RF_clf / rws

np.fill_diagonal(norm, 0)
plt.matshow(norm, cmap=plt.cm.gray)
plt.title('Error_Matrix_RF_Clf')
plt.show()
```

##### (2) Execute principal components analysis (PCA) on the full set of 70,000, generating principal components that represent 95 percent of the variability in the explanatory variables. The number of principal components in the solution should be substantially fewer than the 784 explanatory variables. Record the time it takes to identify the principal components.

```python
split = 60000
X_train, X_test = mnist["data"][:split], mnist["data"][split:]
y_train, y_test = mnist["target"][:split], mnist["target"][split:]
```

```python
# PRINCIPAL COMPONENT ANALYSIS: 95% of variablity

#Record start time of PCA processing

#Set PCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler 

#combine the training and test sets!
X2 = np.concatenate((X_train, X_test), axis=0)  
import time
start = time.perf_counter()
X2 = StandardScaler().fit_transform(X2)   #standardize the data

pca = PCA(n_components=0.95)   #the amount of variance that needs to be explained is >= 95%

pca.fit(X2)  #perform PCA

totimages = pca.transform(X2)   #Apply dimensionality reduction 

print(pca.n_components_, ' principal components can explain 95% of the variance')
#Record end time of PCA processing
end = time.perf_counter()
time = (end - start)
timePCA1 = time 

print("PCA runtime = "+str(timePCA1))    #get the running time for PCA
print(len(X2))
len(totimages)
```

```python
# Using the identified principal components from step (2), 
# train = pd.read_csv('.//digit-recognizer//train.csv') #read in train data
# test = pd.read_csv('.//digit-recognizer//test.csv') #read in test data
# train.shape
# test.shape

# X_train = train.drop('label', 1)  
y_train = train["label"]

# split training and test data
X_trainPC = totimages[0:42000, :]  
print(X_trainPC.shape)
X_testPC = totimages[42000:70000, :]  
print(X_testPC.shape)
#####################################################
# RANDOM FOREST on Dimension Reduced Data (PCA:95% variability)
import time
start = time.perf_counter()

RF_clf = RandomForestClassifier(
  bootstrap = True,
  n_estimators=10,
  max_features='sqrt', 
  random_state=RANDOM_SEED
)

#RANDOM FOREST with MNIST data set (w/ PCA)
RF_clf.fit(X_trainPC, y_train)

#Elapsed Time to perform Random Forest
end = time.perf_counter()
time = (end - start)
timeRF2_42000 = time 
print('Runtime for Random Forest:', timeRF2_42000)
#####################################################
prediction = RF_clf.predict(X_testPC)
print(prediction[0:10])
df=pd.read_csv('.//digit-recognizer//sample_submission.csv') 
df.head(3)
df=df.drop(columns='Label')
df.head(3)
df['Label'] = prediction
df.head(3)
df.to_csv("step2_3_RFPCA_rf_submission2_split42_28_PCA_TRAINING_AND_TEST.csv", index=False)
df.tail()
```

From a management perspective, the predictive accuracy of models must be weighed against the costs of model development and implementation. As  manager of a data science team responsible for implementing models for computer vision (classification of images analogous to the MINST problem), we are tasked with making a recommendation about whether or not to use PCA as a preliminary to machine learning classification.

(2) Discussion of the research design, measurement and statistical methods, traditional and machine learning methods employed 

In this assignment, we used the MNIST dataset to train models to correctly identify digits from handwritten images. According to the kaggle competition description, MNIST ("Modified National Institute of Standards and Technology") is the de facto “hello world” dataset of computer vision. Since its release in 1999, this classic dataset of handwritten images has served as the basis for benchmarking classification algorithms.

We first fitted a random forest classifier using the full set of 784 explanatory variables and the model development set of 60,000 observations (42,000 observations to submit to the kaggle competition). The kaggle score was based on accuracy, and for the 60,000/10,000 dataset, we assessed classification using the F1 score, which is the harmonic mean of precision and recall. Next, we did principal components analysis (PCA) on the full set of 70,000, generating principal components that represent 95 percent of the variability in the explanatory variables. The number of principal components in the solution was substantially fewer than the 784 explanatory variables. Using the identified principal components from step (2), we useD the first 60,000 observations (and 42,000 for kaggle) to build another random forest classifier. Again, kaggle evaluated the submission based on prediction accuracy, and we used the F1 score for the  60,000/10,000 dataset. Our PCA approach was flawed because we should ONLY use the training data for PCA. The experiment was rerun, this time not using the full dataset for PCA, and another random forest classifier was fitted. At the very end, we decided to run one more random forest classifer, this time increasing the number of estimators from 10 to 200. 

---

(3) Overview of programming work; 

We worked with two different versions of the dataset; one from kaggle that was already split into 42,000 training observations and 28,000 test observations; and one from openML, which we split into 60,000 training and 10,000 test observations. This dataset also contained labels for the test observations so we could compute F1 score. We didn't need to do any data prep for the initial random forest classifier, but for the PCA, the data was standardized using StandardScaler. The test set also had to be transformed in order to later make predictions and evaluate the second random forest model. The third version of the random forest classifer was properly done using prinicipal components from the PCA with training data only. The last random forest classifier was run to see if increasing estimators also increased model performance and at what cost (in terms of time).

(4) Review of results with recommendations for management.

The first random forest classifier scored on kaggle achieved an accuracy score of 0.94214, and took about 7 seconds to run, whereas the PCA + random forest classifier (properly done PCA on only training data) achieved an accuracy score of 0.86914, and took 59 seconds to run. Although PCA to reduce dimensionality is a technique to speed up machine learning algorithms, in our case, we did not see this efficiency. Our PCA+RF model took longer to run and had a lower accuracy score than just the RF by itself. Therefore, we would NOT recommend PCA as a preliminary step to machine learning classification in this case. For the kaggle dataset, the PCA+RF actually had lower predictive accuracy and higher costs of model development/implementation. 

Similarly, for the 60,000/10,000 split dataset, the first RF model had an F1 score of 0.9027042584259733, and runtime was 35 seconds. The RF+PCA (train data only) had an F1 score of 0.8023437447548124, and runtime of 109 seconds. Since we saw this lowered predictive accuracy, we attempted to modify the PCA+RF parameters and increased estimatores to 200. The F1 score for this last model was 0.9109481911710775, however, runtime was 780 seconds, or 13 minutes! Again, we see even more clearly that it's better to recommend the standalone RF model instead of PCA+RF, because the predictive accuracy outweighs the high costs of model development/implementation.
